```python
#step3 - refine data
# start with defining good and bad reviews
# let us define good reviews as reviews with ratings of 4 stars and above
# similarly, let us define bad reviews as reviews with ratings of 2 stars and below
bad_reviews = df[(df.stars <= 2 )].text
good_reviews = df[(df.stars >= 4 )].text
# now we have labelled data
```

```python
#step4 - explore text data

# function to tokenize our reviews - to tokenise basically means to split the string into an array of strings, additionally we are also removing common words called stopwords from our reviews as they can skew our data
def tokenize(s):
    word_list = re.findall(r'\w+', s.lower())
    filtered_words = [word for word in word_list if word not in stopwords.words('english')]
    return filtered_words


#n-grams are continous sequences of n-number of words
def count_ngrams(lines, min_length=2, max_length=4):
    lengths = range(min_length, max_length + 1)
    ngrams = {length: collections.Counter() for length in lengths}
    queue = collections.deque(maxlen=max_length)

    # Helper function to add n-grams at start of current queue to dict
    def add_queue():
        current = tuple(queue)
        for length in lengths:
            if len(current) >= length:
                ngrams[length][current[:length]] += 1

    # Loop through all lines and words and add n-grams to dict
    for line in lines:
        for word in tokenize(line):
            queue.append(word)
            if len(queue) >= max_length:
                add_queue()

    # Make sure we get the n-grams at the tail end of the queue
    while len(queue) > min_length:
        queue.popleft()
        add_queue()

    return ngrams


def print_most_frequent(ngrams, num=10):
    for n in sorted(ngrams):
        print('----- {} most common {}-word phrase -----'.format(num, n))
        for gram, count in ngrams[n].most_common(num):
            print('{0}: {1}'.format(' '.join(gram), count))
        print('')
        

def print_word_cloud(ngrams, num=5):
    """Print word cloud image plot """
    words = []
    for n in sorted(ngrams):
        for gram, count in ngrams[n].most_common(num):
            s = ' '.join(gram)
            words.append(s)
            
    cloud = WordCloud(width=1440, height= 1080,max_words= 200).generate(' '.join(words))
    plt.figure(figsize=(20, 15))
    plt.imshow(cloud)
    plt.axis('off');
    plt.show()
    print('')
```

```python
#word cloud of 10 most frequent n-grams in bad and good reviews
most_frequent_bad_reviews = count_ngrams(bad_reviews,max_length=3)
print_word_cloud(most_frequent_bad_reviews, 10)
most_frequent_good_reviews = count_ngrams(good_reviews,max_length=3)
print_word_cloud(most_frequent_good_reviews, 10)
```

```python
#step5 - preprocess data before training
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
stop_words = set(stopwords.words('english')) 

def tokenise(data):
    new_data = []
    for review in data:
        new_review = word_tokenize(review)
        new_review = [i for i in new_review if i not in stop_words]
        new_data.append(new_review)
    return new_data


def lemmatise(data):
    lmtzr = WordNetLemmatizer()
    new_data = []
    for review in data:
        new_review = []
        for word in review:
            new_review.append(lmtzr.lemmatize(word,'v'))
        new_data.append(new_review)
    return new_data

def to_lower(data):
    new_data = []
    for review in data:
        new_review = []
        for word in review:
            new_review.append(word.lower())
        new_data.append(new_review)
    return new_data

def remove_puncs(data):
    new_data = []
    for review in data:
        new_review = []
        for word in review:
            if(word not in string.punctuation):
                new_review.append(word)
        new_data.append(new_review)
    return new_data

def preprocess(data):
    data = tokenise(data)
    data = lemmatise(data)
    data = to_lower(data)
    data = remove_puncs(data)
    return data
            
           
good_reviews_clean = preprocess(good_reviews)
bad_reviews_clean = preprocess(bad_reviews)
print(len(good_reviews_clean),len(bad_reviews_clean))

```
